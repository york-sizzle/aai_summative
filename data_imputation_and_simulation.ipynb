```python
import pandas as pd
import numpy as np
import scipy as sp
import os


data_folder = "./processed_data/"
```

```python
data = pd.read_excel(os.path.join(data_folder, "combined_data.xlsx"), index_col="Year")
data.drop(["pollinating_wild_bees_mean_occupancy", "pollinating_hoverflies_mean_occupancy"], axis=1, inplace=True)

noise_fraction=0.1
# Drop years before 1980
data = data[data.index >= 1980]
cols_to_impute = [col for col in data.columns]# if "plant" not in col]

missing_mask = data[cols_to_impute].isna()
interpolated_data = data[cols_to_impute].interpolate(
    method='polynomial', 
    limit_direction='both',
    fill_value='extrapolate',
    order=1,
    )

# #  Calculate a realistic noise level based on the variance of the interpolated data
noise_std = interpolated_data.std() * noise_fraction

np.random.seed(42)
noise = np.random.normal(0, noise_std, size=interpolated_data.shape)

interpolated_data_with_noise = interpolated_data + noise
data[cols_to_impute] = interpolated_data_with_noise

# Simulate applicable values using a 1st degree polynomial interpolation, which is essentially linear interpolation, but allows for extrapolation at the ends of the data range.
# data[["mean_connectivity_woodland", "mean_connectivity_grassland"]] = data[["mean_connectivity_woodland", "mean_connectivity_grassland"]].interpolate(method="polynomial", order=1, limit_direction="both", fill_value="extrapolate")

# # Fill the gaps without extrapolation for all other columns
# data.interpolate(method="polynomial", order=1, inplace=True)

# # save the processed data
data.to_excel(os.path.join(data_folder, "fully_noise_imputed_data.xlsx"))
data
```
