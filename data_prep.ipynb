```python
import os
import pandas as pd

data_folder = "./data"
files = os.listdir(data_folder)
habitat_connectivity = pd.read_excel(os.path.join(data_folder, 'UK-BDI-2025-habitat-connectivity.xlsx'), sheet_name=None)
insects_wider_countryside = pd.read_excel(os.path.join(data_folder, 'UK-BDI-2025-insects-wider-countryside.xlsx'), sheet_name=None)
plants_wider_countryside = pd.read_excel(os.path.join(data_folder, 'UK-BDI-2025-plants-wider-countryside.xlsx'), sheet_name=None)
pollinating_insects = pd.read_excel(os.path.join(data_folder, 'UK-BDI-2025-pollinating-insects.xlsx'), sheet_name=None)
```

```python
# Find the time range
# extract dates from all dataframes using regex
import re
def extract_dates(df):
    date_pattern = r'(\d{4})'
    dates = []
    for col in df.columns:
        matches = re.findall(date_pattern, col)
        dates.extend(matches)
    return dates

all_dates = set()
earliest_date = float('inf')
latest_date = float('-inf')

for sheet in habitat_connectivity.values():
    all_dates.update(extract_dates(sheet))

for sheet in insects_wider_countryside.values():
    all_dates.update(extract_dates(sheet))

for sheet in plants_wider_countryside.values():
    all_dates.update(extract_dates(sheet))

for sheet in pollinating_insects.values():
    all_dates.update(extract_dates(sheet))

earliest_date = min(all_dates)
latest_date = max(all_dates)
```

```python
print(f"Time range: {earliest_date} - {latest_date}")
```

```python
# Create dataframe for combined data using the time range
combined_data = pd.DataFrame(index=range(int(earliest_date), int(latest_date) + 1))
```
