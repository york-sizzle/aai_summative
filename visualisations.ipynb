```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Set academic plotting style
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_context("paper", font_scale=1.2)
```

```python
df_raw = pd.read_excel("processed_data/combined_data.xlsx")
```

```python
fig = plt.figure()
plt.plot(df_raw['Year'], df_raw['pollinators_mean_occupancy'], color='darkred', linewidth=2, marker='o', markersize=4)
plt.title('UK Pollinator Mean Occupancy (1980-2024)', fontsize=14)
plt.xlabel('Year')
plt.xticks(range(1976, 2030, 4))  # Set x-ticks every 5 years
plt.ylabel('Occupancy Index')
plt.xlim(1976, 2026)
plt.tight_layout()
plt.savefig('figures/eda_pollinator_trend.png', dpi=300)
plt.show()
```

```python
df_imputed = pd.read_excel("processed_data/noise_imputed_data.xlsx")

fig = plt.figure()
# 1. Missing Data Heatmap (Visualizing the 'Alignment' problem)
red_green_cmap = sns.color_palette(["#68fd68", "#fc6363"])  # Red for missing, Green for present
x_index_labels = range(1,df_imputed.shape[1])
x_labels = df_imputed.columns

missing_data = df_imputed.set_index("Year")

g = sns.heatmap(missing_data.isna().T, cbar=False, cmap=red_green_cmap, xticklabels="auto")
g.set_yticklabels(x_index_labels, rotation=0)
plt.title('Missing Values In Processed Dataset With Noise Imputation\n(Red = Missing)', fontsize=14)
plt.ylabel('Feature Index')
plt.xlabel('Years')
plt.tight_layout()
plt.savefig('figures/eda_missing_data_heatmap_processed_with_noise.png', dpi=300)
plt.show()
```
